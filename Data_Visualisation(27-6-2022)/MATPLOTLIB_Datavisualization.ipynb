```python
#You can use the keyword argument marker to emphasize each point with a specified marker
plt.title("LINE_PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.plot([1,2,5,7,8],[3,6,8,9,10],marker="*")
plt.show()
```

```python
#marker=*,o,.,,,x,X,+,p,d,D,H,V,<,>,^,'1','2','3','4'
plt.title("LINE_PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.plot([1,2,5,7,8],[3,6,8,9,10],marker="x")
plt.show()
```

```python
plt.title("LINE_PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.plot([1,2,5,7,8],[3,6,8,9,10],"o")
plt.show()
```

```python
#You can use also use the shortcut string notation parameter to specify the marker.

#This parameter is also called fmt, and is written with this syntax:

#marker|line|color
#marker=*,o,.,,,x,X,+,p,d,D,H,V,<,>,^,'1','2','3','4'
#line=":","--","-.","-"
#color='r','g','b','c','m','y','k','w'
plt.title("LINE_PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.plot([1,2,5,7,8],[3,6,8,9,10],".-.b")
plt.show()
```

```python
#You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:
plt.title("LINE_PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.plot([1,2,5,7,8],[3,6,8,9,10],".-.b",ms=40)
plt.show() 
```

```python
#You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:
plt.title("LINE_PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.plot([1,2,5,7,8],[3,6,8,9,10],".-.b",ms=40,mec="r")
plt.show() 
```

```python
#You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers:
plt.title("LINE_PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.plot([1,2,5,7,8],[3,6,8,9,10],".-.b",ms=40,mec="r",mfc="g")
plt.show() 
```

```python
#You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:
#linestyle="dashed","dotted","dashdot","none","solid"
#you can use the keyword argument color or the shorter c to set the color of the line:
plt.title("LINE_PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.plot([1,2,5,7,8],[3,6,8,9,10],ls="dashed",c="r",linewidth="5")
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Line",fontdict=font1)
plt.xlabel("X-axis",fontdict=font2)
plt.ylabel("Y-axis",fontdict=font2)
plt.plot([10,20,50,30],[5,10,15,6])
plt.plot([12,14,10],[1,5,25])
plt.show()
```

```python
#You can use the loc parameter in title() to position the title.

#Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

import matplotlib.pyplot as plt
import numpy as np
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Line",fontdict=font1,loc='left')
plt.xlabel("X-axis",fontdict=font2)
plt.ylabel("Y-axis",fontdict=font2)
plt.plot([10,20,50,30],[5,10,15,6])
plt.plot([12,14,10],[1,5,25])
plt.show()
```

```python
#Add grid lines to the graph
import matplotlib.pyplot as plt
import numpy as np
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Line",fontdict=font1)
plt.xlabel("X-axis",fontdict=font2)
plt.ylabel("Y-axis",fontdict=font2)
plt.plot([10,20,50,30],[5,10,15,6])
plt.plot([12,14,10],[1,5,25])
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Line",fontdict=font1)
plt.xlabel("X-axis",fontdict=font2)
plt.ylabel("Y-axis",fontdict=font2)
plt.plot([10,20,50,30],[5,10,15,6])
plt.plot([12,14,10],[1,5,25])
plt.grid(axis="x",color="blue",linestyle="--",linewidth=4.5)
plt.show()
```
