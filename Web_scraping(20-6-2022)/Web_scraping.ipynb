```python
import requests
from bs4 import BeautifulSoup
url="https://www.yellowpages.com/"
obj=requests.get(url)
soup=BeautifulSoup(obj.content,"html.parser")
print(soup.text)
```

```python
url1="https://www.bikewale.com/royalenfield-bikes/"
obj1=requests.get(url1)
obj2=BeautifulSoup(obj1.content,'html.parser')
print(obj2.text)
```

```python
img1=obj2.findAll('div',class_="imageWrapper relative")
print(img1)
```

```python
for i in img1:
    j=i.img['src']
    print(j)
```

```python
links=obj2.findAll('div',class_="bikeDescWrapper")
for i in links:
    j=i.a["href"]
    print(j)
```
